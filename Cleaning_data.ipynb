# 2. Cleaning variables

## 2.1 Through string and type methods

### Variable "Sex"

```python
list(df2["Sex"].unique())

#We replace the nan values for "Unknown"
df2["Sex"]=df2["Sex"].fillna("Unknown")

#We replace the values from the variable "Sex" so that the unique values are: "M", "F" and "Unknown".
df2["Sex"] = df2["Sex"].str.replace("M ", "M")
df2["Sex"] = df2["Sex"].str.replace("lli", "Unknown")
df2["Sex"] = df2["Sex"].str.replace(".", "Unknown")
df2["Sex"] = df2["Sex"].str.replace("N", "Unknown")

list(df2["Sex"].unique())

```

## Variable "Fatal (Y/N)"

```python
#We replace the nan values for "UNKNOWN"
df2["Fatal(Y/N)"]=df2["Fatal(Y/N)"].fillna('UNKNOWN')

#We replace the values from the variable "Sex" so that the unique values are: "Y", "N" and "Unknown".
df2["Sex"] = df2["Sex"].str.replace("M ", "M")
df2["Fatal(Y/N)"] = df2["Fatal(Y/N)"].str.replace(" N", "N")
df2["Fatal(Y/N)"] = df2["Fatal(Y/N)"].str.replace("N ", "N")
df2["Fatal(Y/N)"] = df2["Fatal(Y/N)"].str.replace("y", "Y")
df2["Fatal(Y/N)"] = df2["Fatal(Y/N)"].str.replace("M", "UNKNOWN")
df2["Fatal(Y/N)"] = df2["Fatal(Y/N)"].str.replace("2017", "UNKNOWN")

list(df2["Fatal(Y/N)"].unique())
```

## 2.1 Cleaning variables using Regex &/or functions

## Variable "Country"

```python

#First, we take a look at the unique values that are under this variable.
list(df2["Country"].unique())

# We replace nan values with "Not Classified", so that we can categorize them
df2["Country"] = df2["Country"].fillna("Not Classified")

# We apply the function clean_countries to Country.
#df2["Country"].fillna("Not Classified",inplace = True)
df2['Country'] = df2['Country'].apply(edd.clean_countries)

# We visualize the column. 
list(df2["Country"].unique())
```

```python
new_column_continent = {column: column.replace("Country","Continent") for column in columns}
df2.rename(columns = new_column_continent, inplace=True)
```

## Variable "Activity"

```python
#First, we take a look at the unique values that are under this variable.

list(df2["Activity"].unique())

# We replace nan values with "Unknown", so that we can categorize them
df2['Activity']=df2['Activity'].fillna("Unknown")

#We apply the function clean_activity to the column Activity
df2['Activity'] = df2['Activity'].apply(edd.clean_activity)

#We visualize the final values we are left with.
print(list(df2["Activity"].unique()))
```

## Variable "Species"

```python

# First of all, we visualize the unique values in the column Species.
species_list = df2["Species"].fillna('Unknown').unique()

# Using the function word_list_count we can have an idea of how many times each word is repeated.
# This will help us categorize the species in the next step.
edd.word_list_count(species_list)
```

```python
# We get rid off the nan values 
df2['Species'] = df2['Species'].fillna("Unknown")


# We apply the function clean_species to species.
df2['Species'] = df2['Species'].apply(edd.clean_species)

# We visualize the final values in our list.
list(df2["Species"].unique())
```
